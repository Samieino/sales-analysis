```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv('/content/Sample - Superstore.csv', encoding='latin1')
df.head()
```

```python
print(df.info())
print(df.describe())
print(df.isnull().sum())
```

```python
df = df.dropna()
print(df.isnull().sum())
```

```python
df['Profit per Unit'] = df['Profit'] / df['Quantity']
df['Order Value'] = df['Sales'] * df['Quantity']
print(df[['Order ID', 'Product Name', 'Profit per Unit', 'Order Value']].head())
```

```python
# تحليل المبيعات حسب الفئة
category_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)

# تحليل الأرباح حسب الفئة
category_profit = df.groupby('Category')['Profit'].sum().sort_values(ascending=False)

# رسم بياني للمبيعات حسب الفئة
category_sales_plot = category_sales.plot(kind='bar', color='skyblue', figsize=(10,6))
plt.title('Total Sales by Category')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# رسم بياني للأرباح حسب الفئة
category_profit_plot = category_profit.plot(kind='bar', color='orange', figsize=(10,6))
plt.title('Total Profit by Category')
plt.xlabel('Category')
plt.ylabel('Total Profit')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# تحليل المبيعات حسب المنطقة
region_sales = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)

# تحليل الأرباح حسب المنطقة
region_profit = df.groupby('Region')['Profit'].sum().sort_values(ascending=False)

# رسم بياني للمبيعات حسب المنطقة
region_sales_plot = region_sales.plot(kind='bar', color='lightgreen', figsize=(10,6))
plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# رسم بياني للأرباح حسب المنطقة
region_profit_plot = region_profit.plot(kind='bar', color='red', figsize=(10,6))
plt.title('Total Profit by Region')
plt.xlabel('Region')
plt.ylabel('Total Profit')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# تحليل المبيعات حسب المدينة
city_sales = df.groupby('City')['Sales'].sum().sort_values(ascending=False)

# تحليل الأرباح حسب المدينة
city_profit = df.groupby('City')['Profit'].sum().sort_values(ascending=False)

# رسم بياني للمبيعات حسب المدينة
city_sales_plot = city_sales.plot(kind='bar', color='lightcoral', figsize=(12,6))
plt.title('Total Sales by City')
plt.xlabel('City')
plt.ylabel('Total Sales')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# رسم بياني للأرباح حسب المدينة
city_profit_plot = city_profit.plot(kind='bar', color='yellowgreen', figsize=(12,6))
plt.title('Total Profit by City')
plt.xlabel('City')
plt.ylabel('Total Profit')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
# استخراج السنة والشهر من التاريخ
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month_name()
# تحليل المبيعات حسب السنة والشهر
monthly_sales = df.groupby('Month')['Sales'].sum().sort_values(ascending=True)
# رسم بياني للمبيعات حسب الشهر
monthly_sales_plot = monthly_sales.plot(kind='bar', color='skyblue', figsize=(10,6))
plt.title('Total Sales by Month')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# رسم بياني يوضح العلاقة بين الخصم والأرباح
plt.figure(figsize=(10,6))
sns.scatterplot(x='Discount', y='Profit', data=df, color='purple')
plt.title('Discount vs Profit')
plt.xlabel('Discount')
plt.ylabel('Profit')
plt.tight_layout()
plt.show()
```

```python
# تحليل المبيعات حسب الفئة
print("فئات المنتجات الأكثر مبيعًا:")
print(category_sales)
# تحليل الأرباح حسب الفئة
print("فئات المنتجات الأعلى ربحًا:")
print(category_profit)
# تحليل المبيعات حسب المنطقة
print("المناطق الأعلى مبيعات:")
print(region_sales)
# تحليل الأرباح حسب المنطقة
print("المناطق الأعلى ربحًا:")
print(region_profit)
# تحليل المبيعات حسب المدينة
print("المدن الأعلى مبيعات:")
print(city_sales)
# تحليل الأرباح حسب المدينة
print("المدن الأعلى ربحًا:")
print(city_profit)
```
